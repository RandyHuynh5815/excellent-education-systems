## PISA data general structure


```python
import pandas as pd

df = pd.read_csv('student_subset.csv')  
df.head()
```

#### see different countries


```python

# Get unique values from the 'CNT' column
unique_values = df['CNT'].unique()

# Display the unique values
print((unique_values))
```

### see number of NA's for each column


```python
import pandas as pd
import matplotlib.pyplot as plt


# Count the number of NA values for each column
na_counts = df.isnull().sum()

# Create a bar chart
plt.figure(figsize=(10, 6))
na_counts.plot(kind='bar', color='lightcoral')

# Add labels and title
plt.title('Number of NA Values for Each Column', fontsize=16)
plt.xlabel('Columns', fontsize=12)
plt.ylabel('Number of NA Values', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.tight_layout()
plt.show()
```

#### look at the new dataset country expenditures in education: Government expenditure on education as a percentage of GDP (%)


```python
import pandas as pd

# Load the dataset
df_gdp = pd.read_csv('data_2022.csv')

df_gdp.head()

unique_values = df['CNT'].unique()

# Display the unique values
print((unique_values))
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df_gdp = df_gdp[['CNT', 'percentOfGdpOnEducation']]
# Create a bar chart
plt.figure(figsize=(12, 8))
df_gdp.plot(kind='bar', color='skyblue')

# Add labels and title
plt.title('GDP Spending by Country', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('GDP Spending', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.tight_layout()
plt.show()
```
