```python
import re
import numpy as np
import pandas as pd
```

```python
#paths 
infile  = "/Users/Hat/OneDrive/511- Data Vis/Education Project/excellent-education-systems/data-analysis/data/student_subset.csv"
outfile_csv  = "/Users/Hat/OneDrive/511- Data Vis/Education Project/excellent-education-systems/data-analysis/data/student_subset_cleaned.csv"
```

```python
#load
df = pd.read_csv(infile, dtype=str)
```

```python
# Clean Hisced column
df["HISCED"] = (
    df["HISCED"]
    .astype(str).str.strip()  # ensure strings
    .str.replace("ISCED level", "", case=False, regex=False)  # remove the phrase
)

# Convert "less than ISCED level 1" → 0
df.loc[df["HISCED"].str.contains("less", case=False), "HISCED"] = "0"

# Final: convert numeric-looking strings to integers
df["HISCED"] = pd.to_numeric(df["HISCED"], errors="coerce")
```

```python
# Clean ST001D01T column
df["ST001D01T"] = (
    df["ST001D01T"]
    .astype(str).str.strip()  # ensure strings
    .str.replace("Grade", "", case=False, regex=False)  # remove the phrase
)

# Final: convert numeric-looking strings to integers
df["ST001D01T"] = pd.to_numeric(df["ST001D01T"], errors="coerce")
```
