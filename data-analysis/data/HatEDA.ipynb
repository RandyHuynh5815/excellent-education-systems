```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
sns.set(font_scale=1.1)
```

```python
#paths 
data_csv  = "/Users/Hat/OneDrive/511- Data Vis/Education Project/excellent-education-systems/data-analysis/data/student_subset_cleaned.csv"
```

```python
#load
df = pd.read_csv(data_csv, dtype=str)
```

```python
df.info()
```

```python
df[["Math", "Read", "Science", "ESCS", "HISCED", "HISEI"]].describe()
```

```python
df[["Math", "Read", "Science", "ESCS", "HISCED", "HISEI"]].isna().sum()
```

```python
subset = df[["Math", "Read", "Science", "ESCS", "HISCED", "HISEI"]]

plt.figure(figsize=(8,6))
sns.heatmap(subset.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix: Scores + SES Variables")
plt.show()


```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,4))

sns.scatterplot(data=df, x="ESCS", y="Math", ax=ax[0], alpha=0.3)
sns.scatterplot(data=df, x="ESCS", y="Read", ax=ax[1], alpha=0.3)
sns.scatterplot(data=df, x="ESCS", y="Science", ax=ax[2], alpha=0.3)

ax[0].set_title("ESCS vs Math")
ax[1].set_title("ESCS vs Reading")
ax[2].set_title("ESCS vs Science")

plt.tight_layout()
plt.show()


```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,4))

sns.boxplot(data=df, x="HISCED", y="Math", ax=ax[0])
sns.boxplot(data=df, x="HISCED", y="Read", ax=ax[1])
sns.boxplot(data=df, x="HISCED", y="Science", ax=ax[2])

ax[0].set_title("Math by Parental Education (HISCED)")
ax[1].set_title("Reading by Parental Education (HISCED)")
ax[2].set_title("Science by Parental Education (HISCED)")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,4))

sns.scatterplot(data=df, x="HISEI", y="Math", ax=ax[0], alpha=0.3)
sns.scatterplot(data=df, x="HISEI", y="Read", ax=ax[1], alpha=0.3)
sns.scatterplot(data=df, x="HISEI", y="Science", ax=ax[2], alpha=0.3)

ax[0].set_title("HISEI vs Math")
ax[1].set_title("HISEI vs Reading")
ax[2].set_title("HISEI vs Science")

plt.tight_layout()
plt.show()
```

```python
country_scores = df.groupby("CNT")[["Math","Read","Science"]].mean().sort_values("Math")

plt.figure(figsize=(10,6))
sns.barplot(data=country_scores, x=country_scores.index, y="Math")
plt.xticks(rotation=90)
plt.title("Average Math Score by Country")
plt.show()
```
